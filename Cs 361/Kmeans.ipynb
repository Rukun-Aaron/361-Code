```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
X = np.array([[2, 10], [2, 5], [8, 4], [5, 8], [7, 5], [6, 4], [1, 2], [4, 9]])
point_labels = ['A{}'.format(i+1) for i in range(len(X))]  # Labels for each point

# Plotting the initial data points
plt.scatter(X[:, 0], X[:, 1])
for i, label in enumerate(point_labels):
    plt.annotate(label, (X[i, 0], X[i, 1]), textcoords="offset points", xytext=(0, 10), ha='center')
plt.title("Initial Data Points")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# Set the desired number of clusters and cluster centers
k = 2
cluster_centers = np.array([[4, 9], [7, 5]])

# Fit the k-means model
kmeans = KMeans(n_clusters=k, init=cluster_centers, n_init=1, random_state=42)
kmeans.fit(X)

# Get the updated cluster centers
updated_centers = kmeans.cluster_centers_

# Get the predicted cluster labels for the data points
predicted_labels = kmeans.labels_

# Plotting the updated clusters
plt.scatter(X[:, 0], X[:, 1], c=predicted_labels)
for i, label in enumerate(point_labels):
    plt.annotate(label, (X[i, 0], X[i, 1]), textcoords="offset points", xytext=(0, 10), ha='center')
plt.scatter(updated_centers[:, 0], updated_centers[:, 1], marker='x', color='red', label='Cluster Centers')
plt.title("Updated Clusters with Custom Centers")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
