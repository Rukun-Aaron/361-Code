```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt


```

```python
# Generate sample data
X = np.array([[2, 10], [2, 5], [8, 4], [5, 8], [7, 5], [6, 4], [1, 2], [4, 9]])
y = np.array([0, 0, 1, 0, 1, 1, 0, 0])  # Cluster labels (0 or 1)

# Plotting the initial data points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title("Initial Data Points")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# Set the desired cluster centers
center1 = [4, 9]
center2 = [7, 5]

# Fit the k-NN model with the initial centers
k = 2
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X, y)

# Update the cluster centers in the model
knn._fit_X = np.vstack((X, [center1, center2]))
knn._y = np.append(y, [2, 3])  # Additional labels for the centers

# Predict the cluster labels for the updated centers
centers_labels = knn.predict([center1, center2])
center1_label, center2_label = centers_labels[0], centers_labels[1]

# Plotting the updated clusters
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.scatter(center1[0], center1[1], marker='x', color='red', label='Center 1 (Cluster {})'.format(center1_label))
plt.scatter(center2[0], center2[1], marker='x', color='blue', label='Center 2 (Cluster {})'.format(center2_label))
plt.title("Updated Clusters with Custom Centers")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
